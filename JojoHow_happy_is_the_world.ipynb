The Sustainable Development Solutions Network (SDSN) collects data across the world relating to happiness.  They use this data to rank countries in order of happiness factor.

This is not an exact science but can give food for thought in terms of what factors might have the most impact on a nation's happiness levels.

Data is taken from the Gallup World Poll, so not collected directly by SDSN.  
Countries are grouped by region.  

### The factors included are:
Economy (measured in GDP per Capita)	
Family (support systems) 	
Health (measured by Life Expectancy)	
Freedom (sense of)  	 
Trust (Government Corruption)	
Generosity (charitable inclinations)  	
Dystopia Residual 
*  Dystopic is the theoretical most unhappy country with the lowest levels in all six of the above factors  
*  The Residual measure is a calculated as the average of the six distances from lowest

```
# libraries for visualisations
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np

df = pd.read_excel('https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2015.xlsx?raw=true')
happiness = df
#useing plt.figure to resize the barplot
plt.figure(figsize = (16,6))
sns.barplot(x = 'Region',y='Happiness Score',data = happiness)

plt.xticks(rotation = 90)
plt.title("Happiness Score by Region")
plt.show()
```

This barplot shows that North America, Australia and New Zealand have the highest Happiness Score, Sub-Sahara Afica and Southern Asia are the least happiest ragions with the lowest Happinest Scores.

```
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np

happy_2015 = pd.read_excel('https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2015.xlsx?raw=true')
happy_2016 = pd.read_excel('https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2016.xlsx?raw=true')
happy_2017 = pd.read_excel('https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2016.xlsx?raw=true')
happy_2018 = pd.read_excel('https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2018.xlsx?raw=true')
happy_2019 = pd.read_excel('https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2018.xlsx?raw=true')

happy_2018.rename(columns={'Healthy life expectancy': 'Health (Life Expectancy)'}, inplace=True)

happy_2019.rename(columns={'Healthy life expectancy': 'Health (Life Expectancy)'}, inplace=True)
happy_2019.columns
sns.kdeplot(happy_2015['Health (Life Expectancy)'],color='purple')
sns.kdeplot(happy_2016['Health (Life Expectancy)'],color='blue')
sns.kdeplot(happy_2017['Health (Life Expectancy)'],color='green')
sns.kdeplot(happy_2018['Health (Life Expectancy)'],color='pink')
sns.kdeplot(happy_2019['Health (Life Expectancy)'],color='black')


plt.title('Life Expectancy through the Years',size=20)
plt.figure(figsize=(16,6))
plt.show()
```
