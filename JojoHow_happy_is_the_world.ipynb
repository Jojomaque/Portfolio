(Image by Benjamin Scott, [source](https://www.natureindex.com/news-blog/data-visualization-these-are-the-happiest-countries-world-happiness-report-twenty-nineteen))   

The Sustainable Development Solutions Network (SDSN) collects data across the world relating to happiness.  They use this data to rank countries in order of happiness factor.

This is not an exact science but can give food for thought in terms of what factors might have the most impact on a nation's happiness levels.

Data is taken from the Gallup World Poll, so not collected directly by SDSN.  
Countries are grouped by region.  

### The factors included are:
Economy (measured in GDP per Capita)	
Family (support systems) 	
Health (measured by Life Expectancy)	
Freedom (sense of)  	 
Trust (Government Corruption)	
Generosity (charitable inclinations)  	
Dystopia Residual 
*  Dystopic is the theoretical most unhappy country with the lowest levels in all six of the above factors  
*  The Residual measure is a calculated as the average of the six distances from lowest

```
# libraries for visualisations
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np

df = pd.read_excel('https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2015.xlsx?raw=true')
happiness = df
# I am dropping the Standard Error Column as I do not think it is relevant
happiness = happiness.drop(['Standard Error'], axis = 1)

#Exploring the Seaborn correlation matrix
sns.heatmap(happiness.corr(), annot=True,cmap="Purples",fmt=".2f")


#To add a title and specify fontsize
plt.title("Correlation Map", fontsize=18)

#To change the size
plt.figure(figsize=(11,9))  




plt.show()   # Darkest hue is the most correlated
```

The darkest hue shows the highest positive corelation between variables and the lightest hue shows the highest negative correlation.
 Economy(GDP per Capita), Family and Health appear to be the most important contributors to the total happiness score.

```
# libraries for visualisations
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np

df = pd.read_excel('https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2015.xlsx?raw=true')
happiness = df
#useing plt.figure to resize the barplot
plt.figure(figsize = (16,6))
sns.barplot(x = 'Region',y='Happiness Score',data = happiness)


# to rotate the axis labels, I have been doing this when they have been unreadable or overlapping.

plt.xticks(rotation = 90)
#To add a title to the chart

plt.title("Happiness Score by Region")
#To display the bar plot
plt.show()
```

This barplot shows that North America, Australia and New Zealand have the highest Happiness Score, Sub-Sahara Afica and Southern Asia are the least happiest ragions with the lowest Happinest Scores.
