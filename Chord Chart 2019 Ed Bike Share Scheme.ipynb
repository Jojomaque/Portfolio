```python
# Load libraries
import pandas as pd
import holoviews as hv
from holoviews import opts, dim
from bokeh.sampledata.les_mis import data

hv.extension('bokeh')
hv.output(size=200)
```

```python
# Load Data
january = pd.read_csv('https://data.urbansharing.com/edinburghcyclehire.com/trips/v1/2019/01.csv')
february = pd.read_csv('https://data.urbansharing.com/edinburghcyclehire.com/trips/v1/2019/02.csv')
march = pd.read_csv('https://data.urbansharing.com/edinburghcyclehire.com/trips/v1/2019/03.csv')
april = pd.read_csv('https://data.urbansharing.com/edinburghcyclehire.com/trips/v1/2019/04.csv')
may = pd.read_csv('https://data.urbansharing.com/edinburghcyclehire.com/trips/v1/2019/05.csv')
june = pd.read_csv('https://data.urbansharing.com/edinburghcyclehire.com/trips/v1/2019/06.csv')
```

```python

Ed_Bike_2019 = pd.concat([january, february, march, april, may, june])
Ed_Bike_2019.head()
```

```python
journeys = Ed_Bike_2019[['start_station_name','end_station_name']].value_counts()
journey_count = journeys.to_frame().reset_index()
journey_count.rename(columns = {0:'journeys'}, inplace = True)
journey_count
```
