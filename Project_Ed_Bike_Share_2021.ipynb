```python
# Load libraries
import pandas as pd
import holoviews as hv
from holoviews import opts, dim
from bokeh.sampledata.les_mis import data
import seaborn as sns  # plot handling
import matplotlib.pyplot as plt  # plot handling
import folium

hv.extension('bokeh')
hv.output(size=200)
```

```python
# Load Data
january = pd.read_csv('https://data.urbansharing.com/edinburghcyclehire.com/trips/v1/2021/01.csv')
february = pd.read_csv('https://data.urbansharing.com/edinburghcyclehire.com/trips/v1/2021/02.csv')
march = pd.read_csv('https://data.urbansharing.com/edinburghcyclehire.com/trips/v1/2021/03.csv')
april = pd.read_csv('https://data.urbansharing.com/edinburghcyclehire.com/trips/v1/2021/04.csv')
may = pd.read_csv('https://data.urbansharing.com/edinburghcyclehire.com/trips/v1/2021/05.csv')
june = pd.read_csv('https://data.urbansharing.com/edinburghcyclehire.com/trips/v1/2021/06.csv')
```

```python

Ed_Bike_2021 = pd.concat([january, february, march, april, may, june])
Ed_Bike_2021.tail()
```

```python
Ed_Bike_2021.end_station_name.nunique()
```

```python
journeys = Ed_Bike_2021[['start_station_name','end_station_name']].value_counts()
journey_count = journeys.to_frame().reset_index()
journey_count.rename(columns = {0:'journeys'}, inplace = True)
journey_count
```

```python
top_journey = journey_count[:20]
top_journey
```

```python
names = top_journey.start_station_name + '/' + top_journey.end_station_name
plt.figure(figsize=(12,10))
sns.barplot( x=top_journey.journeys[:60], y=names, orient="h") 
```
